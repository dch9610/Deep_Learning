## package 불러오기

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

---

## 이미지 파일 열기

opencv로도 열수 있지만, shape 순서가 바뀔 때도 있어서 PIL로 여는 것을 개인적으로 선호

```python
path = 'image/dog.jpg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
# 색상이 있기 떄문에 3이 붙음 (RGB)
image.shape
```

---

## 이미지 들여다 보기 

```python
image.shape
```

이미지를 열기 전에 shape 및 min, max를 통해서 이미지의 range 확인 필요 

```python
# RGB 값이기 때문에 255를 넘어서는 안됨
np.min(image), np.max(image)
```

## 그래프로 시각화 하기

```python
# image 데이터를 1차원으로 만듬
# 256개로 나눔
# x축을 256까지
plt.hist(image.ravel(),256,[0,256])
plt.plot()
```

---

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

---
## 이미지 흑백으로 열기

```python
path
```

```python
# 이미지가 흑백으로 열림 
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)
```

```python
# 뒤에 3이 없음 => 흑백으로 변함
image_bw.shape
```

---
## 흑백 이미지 열기

```python
# 2개의 차원일 때는 gray scale로 열림
plt.imshow(image_bw)
plt.show()
```

---
### 다른 색상으로 cmap 표현하기

### gray scale

```python
plt.imshow(image_bw,'gray')
plt.show()
```

---

### RdBu

```python
# 레드와 블루로 표현
plt.imshow(image_bw,'RdBu')
plt.show()
```

---

### jet

```python
# 높은값은 빨간색 낮을수록 파란색
plt.imshow(image_bw,'jet')
plt.show()
```

---
## Colorbar 추가하기

```python
plt.imshow(image_bw,'jet')
# 값을 표현해줌
plt.colorbar()
plt.show()
```

## 이미지 설정

이미지 보기 사이즈 조절

```python
plt.figure(figsize=(10,10))
plt.imshow(image_bw,'gray')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title('dog')
plt.imshow(image_bw,'gray')
plt.show()
```
